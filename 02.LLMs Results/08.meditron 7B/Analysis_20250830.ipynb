```python
df = df.dropna(subset='letra_respuesta')
df = df.reset_index(drop=True)
df.shape
```

########################################################################################################################################################################################################

```python
pd.crosstab(df['correct_answer'], df['letra_respuesta'], normalize='columns')
```

```python
df['is_correct'] = df['correct_answer'] == df['letra_respuesta']

results = df.groupby(['source_file', 'year']).agg(
    count_of_matches=('is_correct', 'sum'),
    total_rows=('is_correct', 'size')
).reset_index()
results['percent_correct'] = (results['count_of_matches'] / results['total_rows']) * 100
results['percent_correct'] = results['percent_correct'].round(2)
# results['performance'] = (results['percent_correct'] >= 52).astype(int)
# results[results['performance'] == 1]
results
```

```python
results['model_name'] = 'meditron 7B'
results.to_pickle('/Users/rodrigocarrillo/Documents/Natural Language Processing Projects/Examen Residentado Peru/02.LLMs Results/08.meditron 7B/_Results.pickle')
```
