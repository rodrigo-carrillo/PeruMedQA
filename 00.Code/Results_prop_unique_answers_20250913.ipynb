### Proportions of correct answered provided by each LLM alone.
For example, proportion of questions that only LLM "X" answered correctly, while all other LLMs provided a wrong answer.

```python
answer_cols = [
    'letra_respuesta_medgemma4bit', 
    'letra_respuesta_biomistral7bdare', 
    'letra_respuesta_mediphi', 
    'letra_respuesta_jslmedllama3', 
    'letra_respuesta_openbiollm8b', 
    'letra_respuesta_meditron7b', 
    'letra_respuesta_medgemma27bit', 
    'letra_respuesta_llama3openbiollm70b', 
    'letra_respuesta_medgemma4bitFT'
]

correctness_df = df_combined[answer_cols].apply(lambda col: col == df_combined['correct_answer'])

unique_correct_counts = {}

for subject_col in answer_cols:
    # A subject is uniquely correct if their answer is True AND the sum of all other correct answers is zero
    others_correctness = correctness_df.drop(columns=subject_col)
    
    is_uniquely_correct = (correctness_df[subject_col] == True) & (others_correctness.sum(axis=1) == 0)
    
    # Count the number of questions where this subject was uniquely correct
    unique_correct_counts[subject_col] = is_uniquely_correct.sum()


# Total number of questions (rows)
total_questions = len(df_combined)

# Calculate the proportion for each subject
unique_correct_proportions = {
    subject: 100*count / total_questions 
    for subject, count in unique_correct_counts.items()
}

print("Number of unique correct answers:")
print(unique_correct_counts)
print("\nProportion of unique correct answers:")
unique_correct_proportions
```
